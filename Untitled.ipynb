```python
import pandas as pd
import numpy as np
```

```python
def descretize( df, col):
    col_id = {}
    for col_val in set(df[col]):
        col_id[col_val] = len(col_id)
    df[col] = df.transform({ col: (lambda x : col_id[x]) })
```

```python
df = pd.read_csv('Google-Playstore.csv')
data = df.to_numpy()
```
